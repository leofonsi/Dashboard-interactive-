```
import plotly.graph_objects as go

years=[2025,2030,2035,2040]
hommes=[205831.00,218480.83,231908.10,246160.56]
femmes=[211832.00,227081.28,243428.33,260952.16]
fig=go.Figure()
fig.add_trace(go.Scatter(x=years,y=hommes,mode="lines+markers",name="Hommes"))
fig.add_trace(go.Scatter(x=years,y=femmes,mode="lines+markers",name="Femmes"))
fig.update_layout(title="Prévision de la population par sexe 2025-2040", xaxis_title="Année", yaxis_title="Population estimée", template="plotly_white")
fig.show()
```
