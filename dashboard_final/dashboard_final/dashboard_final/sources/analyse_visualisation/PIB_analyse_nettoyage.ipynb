```python
import pandas as pd
import plotly.express as px

# Charger les fichiers
pib = pd.read_csv(r"../data/PIBRegCour.csv", sep=";")
pib_hab = pd.read_csv(r"../data/PIBHabCour.csv", sep=";")
croissance = pd.read_csv(r"../data/croissancePIBReg.csv", sep=";")
va = pd.read_csv(r"../data/VACourReg10Guelmim-Oued_Noun.csv", sep=";")

# Nettoyage
for df in [pib, pib_hab, croissance, va]:
    df["Année"] = df["Année"].astype(int)
    df["Valeur"] = pd.to_numeric(df["Valeur"], errors="coerce")

# Résultats principaux
pib_2014 = pib.loc[pib["Année"] == 2014, "Valeur"].iloc[0]
pib_2023 = pib.loc[pib["Année"] == 2023, "Valeur"].iloc[0]
evolution_pib = (pib_2023 / pib_2014 - 1) * 100

pib_hab_2014 = pib_hab.loc[pib_hab["Année"] == 2014, "Valeur"].iloc[0]
pib_hab_2023 = pib_hab.loc[pib_hab["Année"] == 2023, "Valeur"].iloc[0]
evolution_pib_hab = (pib_hab_2023 / pib_hab_2014 - 1) * 100

croissance_moyenne = croissance["Valeur"].mean()

print("Analyse économique de Guelmim-Oued Noun")
print("---------------------------------------")
print(f"PIB régional 2014 : {pib_2014:.2f} millions DH")
print(f"PIB régional 2023 : {pib_2023:.2f} millions DH")
print(f"Évolution du PIB : {evolution_pib:.2f} %")
print(f"PIB par habitant 2014 : {pib_hab_2014:.2f} DH")
print(f"PIB par habitant 2023 : {pib_hab_2023:.2f} DH")
print(f"Évolution du PIB par habitant : {evolution_pib_hab:.2f} %")
print(f"Croissance moyenne du PIB : {croissance_moyenne:.2f} %")

# Top 5 secteurs en 2023
#top_secteurs = (
 #   va[va["Année"] == 2023]
  #  .sort_values("Valeur", ascending=False)
   # .head(5)
#)

secteurs_a_exclure = [
    "Produit intérieur brut de la région",
    "Total Valeur ajoutée au prix de base"
]

top_secteurs = (
    va[(va["Année"] == 2023) & (~va["Secteur"].isin(secteurs_a_exclure))]
    .sort_values("Valeur", ascending=False)
    .head(5)
)
# 1. PIB régional interactif
fig_pib = px.line(
    pib,
    x="Année",
    y="Valeur",
    markers=True,
    title="Évolution du PIB régional - Guelmim-Oued Noun",
    labels={"Valeur": "PIB régional (millions DH)"}
)
fig_pib.show()

# 2. PIB par habitant interactif
fig_pib_hab = px.line(
    pib_hab,
    x="Année",
    y="Valeur",
    markers=True,
    title="Évolution du PIB par habitant - Guelmim-Oued Noun",
    labels={"Valeur": "PIB par habitant (DH)"}
)
fig_pib_hab.show()

# 3. Croissance du PIB interactif
fig_croissance = px.line(
    croissance,
    x="Année",
    y="Valeur",
    title="Croissance du PIB régional - Guelmim-Oued Noun",
    labels={"Valeur": "Taux de croissance (%)"}
)
fig_croissance.show()

# 4. Top 5 secteurs interactif
fig_secteurs = px.bar(
    top_secteurs,
    x="Secteur",
    y="Valeur",
    title="Top 5 des secteurs selon la valeur ajoutée en 2023",
    labels={"Valeur": "Valeur ajoutée (millions DH)"}
)

fig_secteurs.update_layout(
    width=1200,   # largeur
    height=700,   # hauteur
    xaxis_tickangle=-45
)

fig_secteurs.show()
```
