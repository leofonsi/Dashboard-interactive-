```python
import pandas as pd
import plotly.express as px

# 1. Charger le fichier
#df["indicateur"] = df["indicateur"].fillna("Non précisé")
df = pd.read_csv(r"../data/fusion_migrations_long.csv")
df["indicateur"] = df["indicateur"].fillna("Immigrants internationaux")

# 2. Aperçu rapide
print(df.head())

# 3. Graphique des indicateurs (barres horizontales)
fig1 = px.bar(
    df,
    x="valeur_numerique",
    y="indicateur",
    color="theme",
    orientation="h",
    title="Analyse des indicateurs de migration"
)

fig1.update_layout(yaxis={'categoryorder':'total ascending'})
fig1.show()


# =====================================================
# Immigrants internationaux et indice d'entrée
# =====================================================

indicateurs_selectionnes = [
    "Immigrants internationaux",
    "Indice d'entrées"
]

df_selection = df[df["indicateur"].isin(indicateurs_selectionnes)].round(2)

fig = px.bar(
    df_selection,
    x="indicateur",
    y="valeur_numerique",
    color="indicateur",
    text="valeur_numerique",
    title="Immigrants internationaux et indice d'entrée"
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Indicateur",
    yaxis_title="Valeur"
)

fig.show()


# 4. Somme des valeurs par thème
df_theme = df.groupby("theme")["valeur_numerique"].sum().reset_index().round(2)

fig2 = px.bar(
    df_theme,
    x="theme",
    y="valeur_numerique",
    color="theme",
    title="Total des valeurs par thème"
)

fig2.show()

df.to_csv(r"../data/outputs/analyse_immigrations.csv", index=False)

```
