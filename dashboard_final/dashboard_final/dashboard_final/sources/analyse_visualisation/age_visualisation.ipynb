```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# =========================
# 1. Chargement du fichier
# =========================
file_path = "../data/RGPH2024-Indicateurs_demographiques_region10.xlsx"

df = pd.read_excel(file_path, sheet_name="Population", header=[0, 1])

labels_row = df.iloc[0]
data = df.iloc[1]

# =========================
# 2. Indicateurs sexe
# =========================
hommes_pct = data[("Sexe : Ensemble", "Sexe (%)")]
femmes_pct = data[("Sexe : Ensemble", "Sexe (%).1")]

hommes_pop = data[("Sexe : Masculin", "Population municipale")]
femmes_pop = data[("Sexe : Féminin", "Population municipale")]

# =========================
# 3. Indicateurs de fécondité
# =========================
fecondite_conjoncturelle = data[("Sexe : Ensemble", "Indicateur conjoncturel de fécondité")]
descendance_finale = data[("Sexe : Ensemble", "Descendance finale des femmes")]

print("Population masculine :", hommes_pop)
print("Population féminine :", femmes_pop)
print("Pourcentage hommes :", hommes_pct, "%")
print("Pourcentage femmes :", femmes_pct, "%")
print("Indicateur conjoncturel de fécondité :", fecondite_conjoncturelle)
print("Descendance finale des femmes :", descendance_finale)

# =========================
# 4. Pyramide des âges
# =========================
groupes_age = labels_row.iloc[69:85].tolist()

hommes_age_pct = data.iloc[69:85].astype(float)
femmes_age_pct = data.iloc[130:146].astype(float)

fig_pyramide = go.Figure()

fig_pyramide.add_trace(go.Bar(
    y=groupes_age,
    x=-hommes_age_pct,
    name="Hommes",
    orientation="h",
    text=hommes_age_pct.round(2),
    textposition="auto"
))

fig_pyramide.add_trace(go.Bar(
    y=groupes_age,
    x=femmes_age_pct,
    name="Femmes",
    orientation="h",
    text=femmes_age_pct.round(2),
    textposition="auto"
))

fig_pyramide.update_layout(
    title="Pyramide des âges - Région Guelmim-Oued Noun RGPH 2024",
    xaxis_title="Pourcentage de la population (%)",
    yaxis_title="Groupes d'âge",
    barmode="relative",
    template="plotly_white",
    height=650
)

fig_pyramide.update_xaxes(
    tickvals=[-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10],
    ticktext=["10", "8", "6", "4", "2", "0", "2", "4", "6", "8", "10"]
)

fig_pyramide.show()

# =========================
# 5. Répartition par sexe
# =========================
df_sexe = pd.DataFrame({
    "Sexe": ["Hommes", "Femmes"],
    "Pourcentage": [hommes_pct, femmes_pct],
    "Population": [hommes_pop, femmes_pop]
})

fig_sexe = px.pie(
    df_sexe,
    names="Sexe",
    values="Pourcentage",
    hole=0.45,
    title="Répartition de la population par sexe"
)

fig_sexe.update_traces(
    textinfo="label+percent",
    hovertemplate="<b>%{label}</b><br>Pourcentage : %{value:.2f}%<extra></extra>"
)

fig_sexe.update_layout(
    template="plotly_white",
    height=500
)

fig_sexe.show()

# =========================
# 6. Indicateurs de fécondité
# =========================
df_fecondite = pd.DataFrame({
    "Indicateur": [
        "Indicateur conjoncturel de fécondité",
        "Descendance finale des femmes"
    ],
    "Valeur": [
        fecondite_conjoncturelle,
        descendance_finale
    ]
})

fig_fecondite = px.bar(
    df_fecondite,
    x="Indicateur",
    y="Valeur",
    text="Valeur",
    title="Indicateurs de fécondité"
)

fig_fecondite.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

fig_fecondite.update_layout(
    yaxis_title="Nombre moyen d'enfants par femme",
    xaxis_title="Indicateur",
    template="plotly_white",
    height=500
)

fig_fecondite.show()
```
